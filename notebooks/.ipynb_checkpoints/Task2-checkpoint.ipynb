```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
time_seq = []
with open('../time/time_seq2.txt') as f:
    lines = f.readlines()
    for line in lines:
        float_line = list(map(float, line.split()))
        time_seq.append(np.mean(float_line))
```

```python
time_par = []
num_p = []
with open("../time/task2.txt") as f:
    lines = f.readlines()
    i = -1
    for line in lines:
        if len(line.split()) == 1:
            i += 1
            num_p.append(int(line))
            time_par.append([])
        else:
            float_line = list(map(float, line.split()))
            time_par[i].append(
                np.mean(float_line)
            )
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 4))



for i, label in zip(range(0, 3), ['1000', '1200', '1400']):
    ax[i].plot(num_p, [time[i] for time in time_par])
    ax[i].scatter(num_p, [time[i] for time in time_par], c='red', marker='s', s=70)
    ax[i].set_xlabel('P')
    ax[i].set_ylabel('Time')
    ax[i].set_title('Matrix size is ' + label)
    ax[i].grid()

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 4))



for i, label in zip(range(0, 3), ['1600', '1800', '2000']):
    ax[i].plot(num_p, [time[i + 3] for time in time_par])
    ax[i].scatter(num_p, [time[i + 3] for time in time_par], c='red', marker='s', s=70)
    ax[i].set_xlabel('P')
    ax[i].set_ylabel('Time')
    ax[i].set_title('Matrix size is ' + label)
    ax[i].grid()

plt.tight_layout()
plt.show()
```

```python
speedup = []
for time_p in time_par:
    speedup.append([time_seq[i] / time_p[i] for i in range(0, 6)])
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 4))



for i, label in zip(range(0, 3), ['1000', '1200', '1400']):
    ax[i].plot(num_p, [speed[i] for speed in speedup])
    ax[i].scatter(num_p, [speed[i] for speed in speedup], c='red', marker='s', s=70)
    ax[i].set_xlabel('P')
    ax[i].set_ylabel('Speedup')
    ax[i].set_title('Matrix size is ' + label)
    ax[i].grid()

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 4))



for i, label in zip(range(0, 3), ['1600', '1800', '2000']):
    ax[i].plot(num_p, [speed[i + 3] for speed in speedup])
    ax[i].scatter(num_p, [speed[i + 3] for speed in speedup], c='red', marker='s', s=70)
    ax[i].set_xlabel('P')
    ax[i].set_ylabel('Speedup')
    ax[i].set_title('Matrix size is ' + label)
    ax[i].grid()

plt.tight_layout()
plt.show()
```

```python
efficiency = []
for j, time_p in enumerate(time_par):
    efficiency.append([time_seq[i] / (time_p[i] * num_p[j]) for i in range(0, 6)])
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 4))



for i, label in zip(range(0, 3), ['1000', '1200', '1400']):
    ax[i].plot(num_p, [eff[i] for eff in efficiency])
    ax[i].scatter(num_p, [eff[i] for eff in efficiency], c='red', marker='s', s=70)
    ax[i].set_xlabel('P')
    ax[i].set_ylabel('Efficiency')
    ax[i].set_title('Matrix size is ' + label)
    ax[i].grid()
plt.tight_layout()
plt.show()
```
